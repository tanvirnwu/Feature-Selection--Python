# Feature Selection

Feature selection is a process in machine learning and data analysis that involves selecting a subset of relevant features (or variables) from a larger set of available features. The goal of feature selection is to identify the most informative and discriminative features that contribute the most to the predictive performance of a model while reducing dimensionality and avoiding overfitting.

```python
import pandas as pd
```

```python
mydata = pd.DataFrame({"A":[1,2,4,1,2,4], 
                    "B":[4,5,6,7,8,9], 
                    "C":[0,0,0,0,0,0],
                    "D":[1,1,1,1,1,1]}) 
```

## Variance Threshold
Variance threshold is a simple feature selection method that is based on the variability of features in a dataset. It is commonly used to remove features with low variance, as they are often considered less informative or redundant.

This feature selection algorithm looks only at the features (X), not the desired outputs (y), and can thus be used for unsupervised learning.

### Applying the same technique on a large dataset

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
df1 = pd.read_csv('E:\Feature Selection\Datasets\santander.csv',nrows=10000)
```

```python
df1.shape
```

```python
X = df1.drop(labels=['TARGET'], axis=1)
y = df1['TARGET']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X, y,test_size=0.3, random_state=0)
X_train.shape, X_test.shape
```

### Applying the variance threshold

```python
var_thres = VarianceThreshold (threshold = 0)
var_thres.fit(X_train)
```

```python
var_thres.get_support()
```

```python
var_thres.get_support().sum()
```

```python
# Number of non-constant features/columns 
len(X_train.columns[var_thres.get_support()])
```

```python
#Number of constant features/columns
constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
for column in constant_columns:
    print(column)
```

```python
X_train.drop(constant_columns,axis=1)
```
