### Check Duplicate Columns

```python
def check_dup_col(df):
    duplicate_columns = set()
    for x in range(df.shape[1]):
        col = df.iloc[:, x]
        for y in range(x + 1, df.shape[1]):
            other_col = df.iloc[:, y]
            if col.equals(other_col):
                duplicate_columns.add(df.columns.values[y])
    
    # Print the duplicate columns
    if duplicate_columns:
        print("Duplicate columns found:")
        print(duplicate_columns)
    else:
        print("No duplicate columns found.")          
```

```python
check_dup_col(df)
```

### Univariate Selection


Univariate Selection, also known as Univariate Feature Selection, is a technique used in machine learning and feature selection to identify the most significant features or variables that contribute the most to the target variable. It involves evaluating each feature independently based on certain statistical measures and selecting the features with the highest scores.

The basic idea behind Univariate Selection is to assess the relationship between each feature and the target variable individually, without considering the interactions or dependencies among features. It assumes that each feature is independent of the others and evaluates its predictive power solely based on its relationship with the target variable.

```python
X = df.iloc[:, :-1]
y = df[['price_range']]
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df.shape
```

```python
### Apply SelectKBest Algorithm
ordered_rank_features= SelectKBest (score_func = chi2, k = 20)
ordered_feature = ordered_rank_features.fit(X,y)
```

```python
ordered_feature.scores_
```

```python
dfscores = pd.DataFrame(ordered_feature.scores_, columns = ["Score"])
dfcolumns = pd.DataFrame(X.columns)
```

```python
features_rank = pd.concat([dfcolumns, dfscores], axis= 1)
features_rank.columns = ['Features', 'Score']
features_rank
```

```python
features_rank.nlargest(10, 'Score')
```
